```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data_xls =  pd.read_excel('G:\MTech\Assignements\Data Mining\Bank Data for case study assignment.xlsx', 'Data')
```

```python
#before droping coloumns with na
data_xls.describe(include ="all")
```

```python
#drop row with atleast one missing value
data_xls.dropna(inplace=True)
```

```python
data_xls.rename(columns={"marital status ": "marital_status", "credit default?": "credit_default"
               , "housing loan?":"housing_loan" , "Personal loan":"Personal_loan"},inplace=True)
```

```python
age_labels = ["{0} - {1}".format(i, i + 9) for i in range(0, 100, 10)]
data_xls["age"] = pd.cut(data_xls.age , range(00,101,10), right=False , labels = age_labels)
```

```python
age = data_xls["age"]
age = age.astype('category').cat.codes
#data_xls["age"].apply(lambda x: x.value_counts()).T.stack()
age.value_counts()
```

```python
data_xls["job"].value_counts(normalize=True)*100
```

```python
data_xls.loc[data_xls['job'] == 'blue-collar', 'job'] = 0
data_xls.loc[data_xls['job'] == 'management', 'job'] = 1
data_xls.loc[data_xls['job'] == 'technician', 'job'] = 2
data_xls.loc[data_xls['job'] == 'admin.', 'job'] = 3
data_xls.loc[data_xls['job'] == 'services', 'job'] = 3
data_xls.loc[data_xls['job'] == 'self-employed', 'job'] = 4
data_xls.loc[data_xls['job'] == 'retired', 'job'] = 4
data_xls.loc[data_xls['job'] == 'entrepreneur', 'job'] = 4
data_xls.loc[data_xls['job'] == 'unemployed', 'job'] = 4
data_xls.loc[data_xls['job'] == 'student', 'job'] = 4
data_xls.loc[data_xls['job'] == 'housemaid', 'job'] = 4
data_xls.loc[data_xls['job'] == 'unknown', 'job'] = 4
data_xls
```

```python
data_xls["marital_status"].value_counts(normalize=True)*100
```

```python
data_xls['age'] = data_xls['age'].astype('category')
data_xls['marital_status'] = data_xls['marital_status'].astype('category')
data_xls['education'] = data_xls['education'].astype('category')
data_xls['credit_default'] = data_xls['credit_default'].astype('category')
data_xls['housing_loan'] = data_xls['housing_loan'].astype('category')
data_xls['Personal_loan'] = data_xls['Personal_loan'].astype('category')
data_xls['y'] = data_xls['y'].astype('category')
cat_columns = data_xls.select_dtypes(['category']).columns
data_xls[cat_columns] = data_xls[cat_columns].apply(lambda x: x.cat.codes)
```

```python
data_xls.head(5)
```

```python
#Sklearn for classification
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score
```

```python
X = data_xls.drop(['y'], axis=1)
y = data_xls['y']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size=0.3, random_state=100)
```

```python
#Train model - Gini
clf = DecisionTreeClassifier(criterion='gini', min_samples_split = 5, min_samples_leaf = 1)
clf.fit(X_train, y_train)
```

```python
y_pred = clf.predict(X_test)
```

```python
acc = accuracy_score(y_test, y_pred)
acc
```

```python
#entropy
clfE = DecisionTreeClassifier(criterion='entropy', min_samples_split = 5, min_samples_leaf = 1)
clfE.fit(X_train, y_train)
```

```python
y_pred = clfE.predict(X_test)
acc = accuracy_score(y_test, y_pred)
acc
```
